Task 1.A. Remove genes with NaNs

```
clean_data = data_paad.dropna()
clean_data.shape
```

Task 1.B. Generate gene expression distribution for all samples

```
trans_data=clean_data.transpose()
```

```
trans_data.shape
```

```
skew=trans_data.skew()
print(skew)
```

For normally distributed data skewness should be close to zero. If skewness>0 then it means that there is more weight on the left tail of the distribution and if skewness<0 then it means that there is more weight on the right tail of the distribution.

```
trans_data.describe()
```

```
cont_features = [cont for cont in list(clean_data.columns)]
clean_data[cont_features].hist(bins=50,figsize=(100,80))
```

```
corr=trans_data.corr()

sns.set(font_scale=1.15)
plt.figure(figsize=(100, 100))

sns.heatmap(corr,square=True, annot=True,cmap='YlGnBu',linecolor="black")
plt.title('Correlation between features');
```
