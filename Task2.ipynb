```python
clean_paad = data_paad.dropna()
clean_paad.shape
```

```python
data= clean_paad.transpose()
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_trans = sc.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca1 = pca.fit_transform(data_trans)
pca_data = pd.DataFrame(data = pca1, columns = ['PCA_1', 'PCA_2'])
pca_data
```

```python
col_paad['histological_type_other'].value_counts()
```

```python
neuroendocrine_type=['neuroendocrine carcinoma nos','neuroendocrine','82463 neuroendocrine carcinoma nos','neuroendocrine carcinoma']
Adenocarcinoma_type=['adenocarcinoma, nos','poorly differentiated pancreatic adenocarcinoma','invasive adenocarcinoma','poorly differentiated adenocarcinoma', 'adenocarcinoma- nos']
        
A=col_paad['histological_type_other'].values

for i, cancer in enumerate(A):
    if cancer in (neuroendocrine_type):
        A[i]='neuroendocrine'
    elif cancer in (Adenocarcinoma_type):
        A[i]='Adenocarcinoma'
               

col_paad['histological_type_other']=A
col_paad['histological_type_other'].value_counts()
```
